```python
import pandas as pd
from pathlib import Path

for file in Path(".").glob("*.csv"):
    print(f"\n{'='*60}")
    print(f"File: {file.name}")
    
    df = pd.read_csv(file)
    
    print(f"Total rows: {df.shape[0]}")
    print(f"Total columns: {df.shape[1]}")
    
    print("\nColumn names:")
    print(df.columns.tolist())
    
    print("\nNull values:")
    print(df.isnull().sum())
    
    print("\nColumn types:")
    print(df.dtypes)
    
    print(f"\nDuplicate rows: {df.duplicated().sum()}")
    
    print("\nMessy / weird values check:")
    for col in df.columns:
        if df[col].dtype == "object":
            print(f"\nColumn: {col}")
            print("Unique sample values:")
            print(df[col].dropna().astype(str).str.strip().value_counts().head(10))
```

```python
import pandas as pd

ab_test = pd.read_csv('ab_test.csv')

initial_count = len(ab_test)

ab_test_cleaned = ab_test.drop_duplicates()

removed_count = initial_count - len(ab_test_cleaned)
print(f"Removed {removed_count} duplicate rows.")

ab_test_cleaned.to_csv('ab_test_cleaned.csv', index=False)
```

```python
import pandas as pd

# Load dashboard file
dashboard = pd.read_csv("dashboard.csv")

# ==============================
# Basic dataset info
# ==============================
print("=" * 60)
print("DASHBOARD DATA CHECK")
print("=" * 60)

print(f"Total rows: {dashboard.shape[0]}")
print(f"Total columns: {dashboard.shape[1]}")

print("\nColumn names:")
print(dashboard.columns.tolist())

# ==============================
# Data type of each column
# ==============================
print("\n" + "=" * 60)
print("COLUMN DATA TYPES")
print("=" * 60)
for col in dashboard.columns:
    print(f"{col} : {dashboard[col].dtype}")

# ==============================
# Column-wise total rows, non-null, null
# ==============================
print("\n" + "=" * 60)
print("COLUMN-WISE DETAILS")
print("=" * 60)
for col in dashboard.columns:
    total_rows = len(dashboard[col])
    non_null = dashboard[col].notnull().sum()
    null_count = dashboard[col].isnull().sum()

    print(f"\nColumn: {col}")
    print(f"Data type: {dashboard[col].dtype}")
    print(f"Total rows: {total_rows}")
    print(f"Non-null values: {non_null}")
    print(f"Null values: {null_count}")

# ==============================
# Check if any null exists
# ==============================
print("\n" + "=" * 60)
print("NULL CHECK")
print("=" * 60)

if dashboard.isnull().values.any():
    print("Yes, there are null values.")
    print("\nNull values per column:")
    print(dashboard.isnull().sum())
else:
    print("No null values found.")

# ==============================
# Check duplicates
# ==============================
print("\n" + "=" * 60)
print("DUPLICATE CHECK")
print("=" * 60)

duplicate_count = dashboard.duplicated().sum()
print(f"Duplicate rows: {duplicate_count}")

if duplicate_count > 0:
    print("Yes, duplicate rows exist.")
else:
    print("No duplicate rows found.")

# ==============================
# Weird / messy values check
# ==============================
print("\n" + "=" * 60)
print("WEIRD / MESSY VALUES CHECK")
print("=" * 60)

for col in dashboard.columns:
    if dashboard[col].dtype == "object":
        print(f"\nColumn: {col}")
        print("Top 10 values:")
        print(
            dashboard[col]
            .dropna()
            .astype(str)
            .str.strip()
            .value_counts()
            .head(10)
        )

# ==============================
# Optional: preview first 7 rows
# ==============================
print("\n" + "=" * 60)
print("FIRST 7 ROWS")
print("=" * 60)
print(dashboard.head(7))
```

```python
import pandas as pd
import os

heatmaps = pd.read_csv("heatmap.csv")

heatmaps["avg_scroll_depth"] = heatmaps["avg_scroll_depth"].round(2)
heatmaps["avg_time_on_page"] = heatmaps["avg_time_on_page"].round(2)
heatmaps["exit_rate"] = heatmaps["exit_rate"].round(2)

heatmaps["page"] = heatmaps["page"].str.strip()

heatmaps["rage_clicks"] = heatmaps["rage_clicks"].astype(int)

os.makedirs("cleaned", exist_ok=True)
heatmaps.to_csv("cleaned/heatmap_cleaned.csv", index=False)

print("Heatmap Cleaning Complete!")
print(heatmaps.head())
```
